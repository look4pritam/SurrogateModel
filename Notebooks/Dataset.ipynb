# Dataset for Surrogate Model

## Import Python packages.

```python
import numpy as np
import pandas as pd
```

## Define design parameters.

```python
n_samples = 1000
n_features = 5
```

## Define bounds for each variable.

```python
bounds = [
    (0, 10),
    (0, 5),
    (-3, 3),
    (1, 7),
    (0, 1)
]
```

## Generate the dataset.

```python
def latin_hypercube_sampling(n_samples, n_features):
    rng = np.random.default_rng()
    result = np.zeros((n_samples, n_features))

    for i in range(n_features):
        perm = rng.permutation(n_samples)
        result[:, i] = (perm + rng.random(n_samples)) / n_samples

    return result
```

### Generate normalized samples.

```python
lhs = latin_hypercube_sampling(n_samples, n_features)
```

### Scale samples to the actual bounds.

```python
X = np.zeros_like(lhs)
for i in range(n_features):
    low, high = bounds[i]
    X[:, i] = low + (high - low) * lhs[:, i]
```

## Define target function.

```python
def target_function(X):
    x1, x2, x3, x4, x5 = X.T

    y = (
        np.sin(x1) +
        np.log(x2 + 1) +
        x3**2 +
        np.sqrt(x4) +
        np.exp(x5)
    )

    return y

y = target_function(X)
```

## Create data frame.

```python
columns = [f"x{i+1}" for i in range(n_features)]
df = pd.DataFrame(X, columns=columns)
df["y"] = y
```

## Save dataset.

```python
df.to_csv("dataset.csv", index=False)
```
